```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('/content/apriori-2.csv')
```

```python
df.head()
```

```python
df = df.set_index('TransactionID')
```

```python
df.dtypes
```

```python
frequent_items = apriori(df,min_support=0.5, use_colnames = True)

rules = association_rules(frequent_items, metric='confidence', min_threshold=1)
rules = rules.sort_values(['confidence','lift'], ascending = [False, False])
rules
```

Dengan Minimum Support sebesar 50% (0.5), terdapat 5 rules yang mendapatkan nilai confident sebesar 100% (1.0). Kesimpulan dari kelima rules tersebut adalah:


*   Jika pembeli membeli (Orange), maka orang tersebut 100% akan membeli (Banana)
*   Jika pembeli membeli (Grape), maka orang tersebut 100% akan membeli (Avocado)
*   Jika pembeli membeli (Avocado), maka orang tersebut 100% akan membeli (Grape)
*   Jika pembeli membeli (Banana, Grape), maka orang tersebut 100% akan membeli (Avocado)
*   Jika pembeli membeli (Banana, Avocado), maka orang tersebut 100% akan membeli (Grape)
